__Description of Data__
1. Data Source & Size:

Source: The dataset for this project was obtained from Kaggle.(https://www.kaggle.com/datasets/chakilamvishwas/imports-exports-15000)

Size: 15,000 records with 16 variables. (2.69 MB)

2. Data Type: This is a Cross-sectional dataset, capturing transactional data at a specific point in time.

3. Data Dimension: Number of Variables: 16
   Number of Observations: Fifteen Thousand
 
4. Data Variable Type:

Text Variables: Transaction_ID, Country, Product, Import_Export, Date, Category, Port, Shipping_Method, Supplier, Customer, Payment_Terms

Number Variables: Integer: Quantity, Customs_Code, Invoice_Number

Decimal: Value ,  Weight

5. Data Variable Category : Identifying Categorical (Nominal & Ordinal) and Non - Categorical Data

These are the columns in the dataset:

- Transaction_ID: A unique identifier for each transaction.  

- Country: The country involved in the transaction.   

- Product: The product being transacted.   

- Import_Export: Whether the transaction is an import or export.   

- Quantity: The number of units involved in the transaction.   

- Value: The monetary value of the transaction.  

- Date: The date the transaction took place.  

- Category: The category of the product. 

- Port: The port through which the transaction occurred. 

- Customs_Code: A code used for customs purposes.  

- Weight: The weight of the goods involved.  

- Shipping_Method: The method used for shipping the goods- land, air, sea).  

 - Supplier: The company or individual supplying the product.  

- Customer: The company or individual eceiving the product.  

 - Invoice_Number: The invoice numer for the transaction.  

 - Payment_Terms: The terms of payment agreed upon for the transaction.

```python
# Import necessary libraries
import pandas as pd
import plotly.express as px

# Group by Payment Method and count the number of transactions for each method
payment_method_totals = sdset.groupby('Payment_Terms').size().reset_index(name='Count')

# Create a pie chart for payment method distribution
fig = px.pie(payment_method_totals, 
             names='Payment_Terms', 
             values='Count', 
             title='Distribution of Transactions by Payment Method',
             color_discrete_sequence=px.colors.qualitative.Plotly)  # Custom color sequence

# Update layout for better visibility with black background and white font
fig.update_layout(
    plot_bgcolor='black',  # Set plot background to black
    paper_bgcolor='black',  # Set outer chart background to black
    font_color='white',     # Set font color to white for readability
    title_x=0.5             # Center the title
)

# Display the pie chart
fig.show()
```

```python
# Import necessary libraries
import pandas as pd
import plotly.express as px


# Filter for both Import and Export data
data_filtered = sdset[sdset['Import_Export'].isin(['Import', 'Export'])]

# Group by Category and sum the import/export values as transaction value
category_totals = data_filtered.groupby('Category').agg({'Value': 'sum'}).reset_index()

# Rename the 'Value' column to 'Transaction Value'
category_totals.rename(columns={'Value': 'Transaction Value'}, inplace=True)

# Sort the data by Transaction Value in descending order
category_totals = category_totals.sort_values(by='Transaction Value', ascending=False)

# Identify the category with the highest transaction value
highest_transaction = category_totals.loc[category_totals['Transaction Value'].idxmax()]

# Print the category with the highest transaction value
print(f"Category with the highest transaction value: {highest_transaction['Category']} - Value: {highest_transaction['Transaction Value']}")

# Create a bar plot showing the total transaction value by category with different colors
fig = px.bar(category_totals, 
             x='Category', 
             y='Transaction Value', 
             title='Total Transaction Value by Category (Descending Order)',
             labels={'Transaction Value': 'Total Transaction Value', 'Category': 'Category'},
             color='Category',  # Use 'Category' for color differentiation
             color_discrete_sequence=px.colors.qualitative.Plotly)  # You can choose any color sequence here

# Update layout for better visibility with black background and white font
fig.update_layout(
    xaxis_title='Category',
    yaxis_title='Total Transaction Value',
    title_x=0.5,  # Center the title
    height=600,   # Adjust height for better visibility
    plot_bgcolor='black',  # Set plot background to black
    paper_bgcolor='black',  # Set outer chart background to black
    font_color='white'      # Set font color to white for readability
)

# Display the bar plot
fig.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Step 3: Select non-categorical variables
# Assuming 'Quantity', 'Value', 'Weight', and 'Date' are the non-categorical variables
non_categorical_vars = sdset[['Quantity', 'Value', 'Weight']]

# Step 4: Calculate correlation matrix
correlation_matrix = non_categorical_vars.corr()

# Step 5: Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Optional: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.8f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import plotly.express as px

# Convert 'Date' column to datetime format if not already done
sdset['Date'] = pd.to_datetime(sdset['Date'], format='%d-%m-%Y')

# Extract the year from the Date
sdset['Year'] = sdset['Date'].dt.year

# Group by 'Year' and calculate the average 'Value' for that year
average_value_per_year = sdset.groupby('Year').agg({'Value': 'mean'}).reset_index()

# Create a line chart
fig = px.line(average_value_per_year, 
              x='Year', 
              y='Value', 
              title='Change in Average Value Over Three Years',
              labels={'Value': 'Average Value', 'Year': 'Year'},
              line_shape='linear')  # Set the line shape

# Update layout for better visibility
fig.update_layout(
    plot_bgcolor='black',  # Set plot background to black
    paper_bgcolor='black',  # Set outer chart background to black
    font_color='white',     # Set font color to white for readability
    title_x=0.5            # Center the title
)

# Display the line chart
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Getting the top 10 categories by frequency or total value
top_10_categories = sdset['Product'].value_counts().nlargest(10).index

# Filtering the DataFrame to only include the top 10 categories
top_10_sample = sdset[sdset['Product'].isin(top_10_categories)]

# Setting the background to black
plt.style.use('dark_background')

# Setting a vibrant color palette
vibrant_palette = sns.color_palette("bright")

# Customizing the outliers (fliers)
flierprops = dict(marker='o', color='red', markersize=6)

# Boxplot for value based on top 10 product categories with vibrant colors and distinct outliers
sns.boxplot(x='Product', y='Value', data=top_10_sample, palette=vibrant_palette, flierprops=flierprops)

# Title and rotation for readability
plt.title('Boxplot of Value by Top 10 Products')
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 5 products by total value
top_5_products = sdset.groupby('Product')['Value'].sum().nlargest(5).index

# Filter the dataset for these top 5 products
top_5_data = sdset[sdset['Product'].isin(top_5_products)]

# Calculate the average weight per product and import/export category
avg_weight_top_5 = top_5_data.groupby(['Product', 'Import_Export'])['Weight'].mean().reset_index()

# Create the grouped bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Weight', hue='Import_Export', data=avg_weight_top_5, palette='viridis')

# Adding titles and labels
plt.title('Average Weight for Top 5 Products by Value Segmented by Import/Export', fontsize=16)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Average Weight', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Import/Export')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Step 1: Group by country and calculate the frequency of transactions
country_frequency = sdset.groupby('Country')['Value'].count().reset_index()
country_frequency.columns = ['Country', 'Transaction_Frequency']

# Step 2: Plotting the geographical heatmap with updated settings
fig = px.choropleth(country_frequency, 
                    locations="Country", 
                    locationmode='country names', 
                    color="Transaction_Frequency", 
                    hover_name="Country", 
                    color_continuous_scale="Blues",  # New color palette
                    title="Geographical Heat Map of Transaction Frequency by Country")

# Step 3: Adjust the layout for map size, title font size, black background, and map background
fig.update_layout(
    title_font_size=20,       # Reduce title font size
    width=1200,               # Increase map width
    height=800,               # Increase map height
    margin={"r":0,"t":50,"l":0,"b":0},  # Reduce space around the map
    paper_bgcolor="black",    # Set overall background color to black
    plot_bgcolor="black",     # Set plot background color to black
    font_color="white"        # Change font color to white for visibility
)

# Step 4: Change the map background color to black
fig.update_geos(
    bgcolor="black",          # Set the map background to black
    showcoastlines=False,     # Optional: remove coastlines for a cleaner look
    showland=True,            # Ensure land is shown
    landcolor="gray"          # Set land color to gray for better visibility
)

# Step 5: Show the figure
fig.show()
```

```python
import plotly.graph_objects as go

# Step 1: Calculate the average transaction value
average_value = sdset['Value'].mean()

# Step 2: Create the gauge indicator with dark brown and dark blue colors, along with a black background
fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=average_value,
    title={'text': "Average Transaction Value", 'font': {'color': 'white'}},
    gauge={
        'axis': {'range': [None, sdset['Value'].max()], 'tickcolor': "white"},
        'bar': {'color': "darkblue"},  # Dark blue for the gauge bar
        'steps': [
            {'range': [0, average_value * 0.5], 'color': "black"},   # Dark brown
            {'range': [average_value * 0.5, average_value], 'color': "black"},  # Lighter brown (still dark)
            {'range': [average_value, sdset['Value'].max()], 'color': "black"}  # Dark slate blue
        ],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': average_value
        },
    },
    number={'font': {'color': 'white'}}
))

# Step 3: Adjust the layout for a black background and dark-themed colors
fig.update_layout(
    paper_bgcolor="black",  # Black background
    plot_bgcolor="black",   # Black plot area
    font={'color': "white"} # White font for text
)

# Step 4: Display the gauge indicator
fig.show()
```
